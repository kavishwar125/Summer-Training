```python
netflix = pd.read_csv('../input/dataset/NFLX.csv')
```

```python
netflix['Date'] = pd.to_datetime(netflix['Date'], format='%Y-%m-%d')
netflix['Date'] = netflix["Date"].values.astype('float64')
netflix.dtypes
```

```python
netflix.describe()
```

```python
plt.hist(netflix.Close,color='Red')
plt.show()
```

```python
target =np.log(netflix.Close)
plt.hist(target,color='Blue')
plt.show()
```

```python
corr=netflix.corr()
print(corr['Close'].sort_values(ascending=False)[:5],'\n')
print(corr['Close'].sort_values(ascending=False)[-5:],'\n')
```

```python
pivot_netflix=netflix.pivot_table(index='High',values='Close', aggfunc=np.median)
pivot_netflix.plot(kind='bar', color='blue')

plt.xlabel('High')
plt.ylabel('Closing Price')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.scatter(x=netflix['High'], y=np.log(netflix['High']))
plt.ylabel('Closing Price')
plt.xlabel('Higher Price')

plt.show()
```

```python
X= netflix[['Date','Open','High','Low','Volume']]
y = netflix['Close']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print('root square error is',regressor.score(X_test, y_test))
predictions=regressor.predict(X_test)

print('RMSE', mean_squared_error(y_test, predictions))
```
